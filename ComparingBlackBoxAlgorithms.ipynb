```python
%matplotlib inline

import time
import numpy as np
from matplotlib import pyplot as pl
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import Axes3D
import pandas

from pybrain.optimization import NelderMead
from pybrain.optimization import CMAES
from pybrain.optimization import HillClimber
from pybrain.optimization import GA
from pybrain.optimization import OriginalNES
from pybrain.optimization import ExactNES
from pybrain.optimization import FEM


class Function(object):
    def __init__(self, name, func, initial, target):
        self.name = name
        self.func = func
        self.initial = initial
        self.target = target
        
    def __call__(self, x):
        return self.func(x)
    
    
def simple_f(x):
    return x[0] * np.sin(x[0])

def branin_f(x):
    a = 1
    b = 5 / (4*np.pi*np.pi)
    c = 5 / np.pi
    d = 6
    e = 10
    f = 1 / (8*np.pi)
    return a * (x[1] - b*x[0]*x[0] + c*x[0] - d)**2 + e * (1-f)*np.cos(x[0]) + e

def goldstein_price_f(x):
    return (1 + (x[0]+x[1]+1)**2 * (19-14*x[0]+3*x[0]**2-14*x[1]+6*x[0]*x[1]+3*x[1]**2)) * \
            (30 + (2*x[0]-3*x[1])**2 * (18-32*x[0]+12*x[0]**2+48*x[1]-36*x[0]*x[1]+27*x[1]**2))

def plot_evaluations(func, num_params, args, value):
    fig = pl.figure(figsize=(10, 10))
    if num_params == 1:
        x = np.atleast_2d(np.linspace(0, 10, 1000)).T

        pl.plot(x, func([x]), 'r:', label=func.name)
        pl.plot(args, value, 'b.', markersize=7,
                label=u'evaluations')
        pl.ylim(-10, 10)
        pl.legend(loc='upper left')
        pl.show()
    elif num_params == 2:
        x, y = np.meshgrid(np.arange(-3., 8., 0.1), np.arange(-3., 8., 0.1))
        zs = np.array([func([i, j]) for i, j in zip(np.ravel(x), np.ravel(y))])
        z = zs.reshape(x.shape)

        X = []
        Y = []
        for el in args:
            X.append(el[0])
            Y.append(el[1])
        sp = fig.add_subplot(111, projection='3d')
        sp.plot_surface(x, y, z, cmap=cm.hot, alpha=0.2)
        sp.scatter(X, Y, value, c='r', zorder=1)
    

def maximize_f(func, num_evaluations, optimizer):
    num_params = len(func.initial)
    x0 = np.array(func.initial)

    #t0 = time.clock()
    l = optimizer(func, x0, minimize=True, storeAllEvaluations=True, storeAllEvaluated=True, 
                  maxEvaluations=num_evaluations)
    out = l.learn()
    #t_diff = time.clock() - t0
    print '----------------------------------------------------------'
    print optimizer.__name__
    if num_params == 1:
        print 'Min: x=', out[0][0], '; y=', out[1]
        print 'Error=', np.abs(out[0] - func.target)[0]
    elif num_params == 2:
        print 'Min: x=', out[0][0], '; y=', out[0][1], '; z=', out[1]
        print 'Error=', np.linalg.norm(out[0] - func.target)
    #print 'Time=', t_diff, 'seconds'

    plot_evaluations(func, num_params, l._allEvaluated, l._allEvaluations)


functions = {
    'simple'         : Function('x*sin(x)', simple_f, [3.], 4.91318),
    'branin'         : Function('Branin', branin_f, [-1., 1.], (np.pi, 2.275)),
    'goldstein-price': Function('Goldstein-Price', goldstein_price_f, [2., 2.], (0., -1.))
}

optimizers = [
    NelderMead, CMAES, HillClimber, GA, OriginalNES, ExactNES, FEM
]
```

```python
maximize_f(functions['branin'], 50, NelderMead)
```

```python
maximize_f(functions['branin'], 50, CMAES)
```

```python
maximize_f(functions['branin'], 200, HillClimber)
```

```python
maximize_f(functions['branin'], 200, GA)
```

```python
maximize_f(functions['branin'], 300, FEM)
```

```python
maximize_f(functions['simple'], 30, HillClimber)
```

```python
maximize_f(functions['simple'], 30, GA)
```

```python
maximize_f(functions['goldstein-price'], 120, OriginalNES)
```

```python
maximize_f(functions['goldstein-price'], 120, ExactNES)
```

```python
maximize_f(functions['goldstein-price'], 120, FEM)
```

```python
maximize_f(functions['goldstein-price'], 120, CMAES)
```
